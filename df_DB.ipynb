```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
```

```python

# File to Load (Remember to change these)
nfldata_csv = "C:/Users/admin/Documents/SMU/NFL-QB-Analytics/Assets/Data_Files/Cleaned_QB_data.csv"

# Read the City and Ride Data
nfldata = pd.read_csv(nfldata_csv)
nfldata.head()
```

```python
nfldata_dt = nfldata.groupby(["Year", "Name"])["Wins"].sum().reset_index().sort_values(by = "Wins",ascending = False)

nfldata_dt
```

```python
nfldata_dt = nfldata.groupby(["Year", "Name"])["Passing Yards"].sum().reset_index().sort_values(by = "Passing Yards",ascending = False)

nfldata_dt
```

```python
nfldata_df = pd.DataFrame(nfldata_dt)
nfldata_df 
```
